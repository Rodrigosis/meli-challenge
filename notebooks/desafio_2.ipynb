```python
import pandas as pd
```

```python
data = pd.read_csv('dataset.csv', names=['personagem_1', 'personagem_2', 'interacoes', 'livro'])
data
```

```python
personagens = pd.DataFrame()
personagens['personagem'] = list(data['personagem_1']) + list(data['personagem_2'])
personagens = personagens.drop_duplicates()
personagens
```

```python
result = {}

for p in personagens.personagem:
    df = data[(data['personagem_1'] == p) | (data['personagem_2'] == p)]

    amigos = list(df['personagem_1']) + list(df['personagem_2'])
    amigos = list(filter(lambda a: a != p, amigos))

    for a_1 in amigos:
        for a_2 in amigos:
            if a_1 == a_2:
                continue

            if a_1 in result:
                if a_2 in result[a_1] and p not in result[a_1][a_2]:
                    result[a_1][a_2].append(p)
            elif a_2 in result:
                if a_1 in result[a_2] and p not in result[a_2][a_1]:
                    result[a_2][a_1].append(p)
            else:
                result[a_1] = {a_2: [p]}

print(result)
```
